```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
import math
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
```

```python
data = load_iris()
X=data.data[:,:2]
y=data.target
X.shape,y.shape
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap='gist_rainbow')
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,random_state=10,test_size=0.25)
```

```python
model = KMeans(n_clusters=3,random_state=21)
```

```python
model.fit(x_train,y_train)
```

```python
center = model.cluster_centers_
center
```

```python
ypred = model.predict(x_test)
ypred
```

```python
new_label = model.labels_
```

```python
plt.scatter(x_train[:,0],x_train[:,1],c=new_label,cmap='gist_rainbow')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],marker='*',s=100)
```

```python
print(model.predict([X[50]]))
```

```python
K=range(1,10)
d=[]
for k in K:
    km = KMeans(n_clusters=k)
    km.fit(x_test,y_test)
    d.append(km.inertia_)
d
```
