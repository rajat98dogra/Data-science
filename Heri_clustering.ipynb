```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import  dendrogram,linkage
```

```python
from sklearn.datasets import load_iris
```

```python
data=load_iris().data
y = load_iris().target
```

```python
sns.scatterplot(x=data[:,0],y=data[:,2],hue=y,data=data)
```

```python
cluster = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='single')
cluster
```

```python
ypred = cluster.fit_predict(data)
```

```python
ypred
```

```python
plt.scatter(x=data[:,0],y=data[:,1],c=y)
plt.figure()
plt.scatter(data[:,0], data[:,1], c=cluster.labels_) 
```

<pre>
‘ward’ minimizes the variance of the clusters being merged.

‘average’ uses the average of the distances of each observation of the two sets.

‘complete’ or ‘maximum’ linkage uses the maximum distances between all observations of the two sets.

‘single’ uses the minimum of the distances between all observations of the two sets.
</pre>


```python
linkages=['ward','complete', 'average', 'single']
for i in linkages:
    cluster = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage=i)
    ypred = cluster.fit_predict(data)
    f_1=0;f_2=2
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.title(f'actual')
    plt.scatter(x=data[:,f_1],y=data[:,f_2],c=y)
    plt.subplot(1,2,2)    
    plt.title(f'{i} predicts')
    plt.scatter(data[:,f_1], data[:,f_2], c=cluster.labels_) 
```

```python
for i in linkages:
    plt.figure(figsize=(10,10))
    plt.title(i)
    dend = dendrogram(linkage(data, method=i))
```
