```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd
import math
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
SIZE=20
np.random.seed(5)
df= pd.DataFrame(columns=['price','rooms','age','area'])
df.area= np.array(np.random.randint(4000,5000,size=(SIZE)))
df.rooms= np.array(np.random.randint(2,5,size=(SIZE)))
df.age= np.array(np.random.randint(5,25,size=(SIZE)))
price= []
# df.shape
for i in range(SIZE):
    price.append(math.floor(df.rooms.iloc[i]*4000 + df.area.iloc[i]%10*1000 + df.age.iloc[i]*0.6 ))
```

```python
df.price=np.array(price)
df
```

```python
plt.figure(figsize=(12,5))
sns.barplot(df.price,df.area)
```

```python
X=df.iloc[:,1:]
y=df.iloc[:,0]
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
ypred=lr.predict(x_test)
ypred
```

```python
residual = np.array(y_test - ypred)
residual
```

```python
plt.plot(residual)
```
