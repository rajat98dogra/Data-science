```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import seaborn as sns
```

```python
M=50
S=12
X = np.random.normal(M,S,200)
```

```python
plt.hist(X)
```

```python
def find_mu(x,mu,sigma):
    return (1/np.sqrt(2*np.pi*(sigma**2)))*(np.exp( - ((x -mu)**2 / (2*sigma**2)) ))
def find_sigma(x,mu,sigma):
    return (1/np.sqrt(2*np.pi*(sigma**2)))*(np.exp( - ((x -mu)**2 / (2*sigma**2)) ))
```

```python
m=[]
for x in X:
    res=float('-inf')
    lik = []
    for mu in range(0,100,10):
        l = find_mu(x,mu,S)
        lik.append(l)
    res = lik.index(max(lik))*10
    m.append(res)
plt.hist(m)
np.unique(m,return_counts=True)
```
